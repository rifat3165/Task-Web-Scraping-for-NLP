```python
poems_contents = []

for i in df5["Poems_Url"]:
    target_url3 = i
    response = requests.get(target_url3)
    html = response.text
    soup = BeautifulSoup(html, "html.parser")
    web_content = soup.find("div", class_="post_description")

    if web_content:
        poem_elements = web_content.find("p")
        if poem_elements:
            poem_content = poem_elements.text.strip()
            poems_contents.append(poem_content)
        else:
            poems_contents.append(" ")
    else:
        poems_contents.append(" ")
```

```python
df5["Poem_Content"] = poems_contents
```

```python
def clean_text(text):
    text = re.sub(r',', '', text)
    text = re.sub(r'|', '', text)
    text = re.sub(r';', '', text)
    text = re.sub(r'!', '', text)
    text = re.sub(r'-', '', text)
    #text = ''.join([c for c in text if ord(c) < 128])
    return text

df5['Poet_Name'] = df5['Poet_Name'].apply(clean_text)
df5['Poem_Name'] = df5['Poem_Name'].apply(clean_text)
df5['Poem_Genre'] = df5['Poem_Genre'].apply(clean_text)
df5['Poem_Content'] = df5['Poem_Content'].apply(clean_text)

df5.to_csv('web_scraping_cleaned_dataset.csv', index=False)
df5.to_csv(r'C:\Data Science and ML\web scraping\web_scraping_cleaned_dataset.csv', index=False)
```
